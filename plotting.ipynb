```python
import matplotlib.pyplot as plt
import pandas as pd

data_df = pd.read_excel(r"D:\Thesis\2024\OE-NMC10Ah-45SOC_ssss.xlsx")  # After interpolation
data_rf = pd.read_excel(r"D:\Thesis\2024\OE-NMC10Ah-45SOC.xlsx")  # Original


data_rf['Time_voltage'] = data_rf['Time (second)'].astype(float)
data_rf['Time_temp'] = data_rf['Time (sec) '].astype(float)

data_rf.dropna(subset=['TC1 (C)'], inplace=True)
data_df.dropna(subset=['Temperature(C)'], inplace=True)

plt.figure(figsize=(12, 8))

# Plotting voltage data
plt.subplot(2, 1, 1)  
plt.plot(data_rf['Time_voltage'], data_rf['Voltage (V)'], label='Original Voltage', color='blue')
plt.plot(data_df['Time(s)'], data_df['Voltage(V)'], label='Interpolated Voltage', color='red', linestyle='--')
plt.xlabel('Time (seconds)')
plt.ylabel('Voltage (V)')
plt.title('Comparison of Original and Interpolated Voltage')
plt.legend()
plt.grid(True)

# Plotting temperature data
plt.subplot(2, 1, 2)  
plt.plot(data_rf['Time_temp'], data_rf['TC1 (C)'], label='Original Temp', color='blue')
plt.plot(data_df['Time(s)'], data_df['Temperature(C)'], label='Interpolated Temp', color='red', linestyle='--')
plt.xlabel('Time (seconds)')
plt.ylabel('Temperature (C)')
plt.title('Comparison of Original and Interpolated Temperature')
plt.legend()
plt.grid(True)

plt.tight_layout()  
plt.show()
```
